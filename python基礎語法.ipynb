````python
import random

# my = int(input("0-5:"))
# com = random.randint(0, 5)
my = 2
com = 0
trans = ["剪刀(0)", "石頭(0)", "布(0)",
      "剪刀(1)", "石頭(1)", "布(1)"]
print("我出的:", trans[my])
print("電腦的:", trans[com])

# 第一步:先確定他們是否同組
# 0 // 3: 0
# 1 // 3: 0
# 2 // 3: 0
# 3 // 3: 1
# 4 // 3: 1
# 5 // 3: 1
# 6 // 3: 2
# 取商: 分組 取餘: 循環
mygroup = my // 3
comgroup = com // 3
if mygroup == comgroup:
      # 情況1: my, com都是0~2(0~2)
      # 情況2: my, com都是3~5(0~2)
      # 如果是3~5 回到0~2(餘數)
    my2 = my % 3
    com2 = com % 3
    if my2 == (com2 + 1) % 3:
      print("win")
    elif my2 == (com2 - 1) % 3:
      print("Lose")
    else:
      print("Even")

else:
    print("平手(不同組)")
````

````python
ord("a")
bin(ord("A"))

s = "dsdfjjfJKLLLK  df565321 198798e12. , [ ]"
#英文大小寫統計
lcase, ucase = 0, 0
i = 0
while i < len(s):
  unit = s[i]
  if ord("a") <= ord(unit) <= ord("z"):
    lcase += 1
  elif ord("A") <= ord(unit) <= ord("Z"):
    ucase += 1
  i += 1
print(lcase,ucase)
````

````python
s = "dsdfjjfJKLLLK  df565321 198798e12. , [ ]"
#英文字母統計
lcases = [0] * 26
ucases = [0] * 26
i = 0
while i < len(s):
  unit = s[i]
  if ord("a") <= ord(unit) <= ord("z"):
    alpha_n = ord(unit) - ord("a")
    lcases[alpha_n] += 1
  elif ord("A") <= ord(unit) <= ord("Z"):
    alpha_n = ord(unit) - ord("A")
    ucases[alpha_n] += 1
  i += 1
print(lcases, ucases)

i = 0
while i < 26:
  strs = chr(ord("a") + i) + ": " + str(lcases[i])
  print(strs)
  i += 1
````

````python
import json
import urllib.request as req

while True:
    url = "https://www.cosmed.com.tw/api/getStore.aspx?t=store&c=&d=&s="
    f = req.urlopen(url)
    # content為bytes形態: 原始的01
    content = f.read()
    # bytes形態有一個叫做decode的專屬功能, 幫妳轉換形態str
    content = content.decode()
    # print(type(content))
    if "Invalid" in content:
        print("retry...")
    else:
        break
# print(type(content))
# 型態轉換: str -> dict/list
content = json.loads(content)
# print(type(content))
# content是字典, 兩組: data, result
store_list = content["data"]
# store是一家店(字典)
for store in store_list:
    print(store["StoreNM"])
    print(store["Tel"])
    print(store["Address"])
    print("-" * 30)
````

````python
import json
import urllib.request as req

while True:
    url = "https://www.cosmed.com.tw/api/getStore.aspx?t=store&c=&d=&s="
    f = req.urlopen(url)
    content = f.read()
    content = content.decode()
    if "Invalid" in content:
        print("retry...")
    else:
        break

# 準備儲存的檔案
f = open("cosmed.csv", "w", encoding="utf-8")
f.write("名稱, 電話\n")
content = json.loads(content)
store_list = content["data"]
for store in store_list:
    print(store["StoreNM"])
    print(store["Tel"])
    print(store["Address"])
    print("-" * 30)
    f.write(store["StoreNM"])
    f.write(",")
    f.write(store["Tel"])
    f.write("\n")

# 關閉
f.close()
````
