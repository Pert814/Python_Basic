````python
import requests
import codecs

# 定義 URL 和請求標頭
url = "https://www.ezlang.net/ajax/tool_data.php"
headers = {
    "Connection": "keep-alive",
    "Content-Type": "application/x-www-form-urlencoded; charset=UTF-8",
    "X-Requested-With": "XMLHttpRequest",
}

# 定義請求數據
payload = {
    "txt": "幸せは自分自身の心の中に見つけることができる",
    "sn": "kana",
}

# 發送 POST 請求
try:
    response = requests.post(url, headers=headers, data=payload)

    # 檢查請求是否成功
    if response.status_code == 200:
        print("成功獲取數據：")
        # print(response.text)  # 打印伺服器返回的數據
    else:
        print(f"請求失敗，狀態碼：{response.status_code}")
except requests.RequestException as e:
    print(f"請求過程中出現錯誤：{e}")

response = codecs.decode(response.text, 'unicode_escape')
text = ""
for i in response:
  if ord("ぁ") <= ord(i) <= ord("゜"):
    text = text + i
print(text)
````

HTML格式

```python
長方形區塊構成, 一個包含一個

<a href="網址"> -> start

</a> -> end

1. 名字: 決定它的功能
  !!! 開始和結束都寫一遍
  a: 超連結
  img/video: 圖片/影片

2. 屬性: 決定這個特別區塊的一些細微屬性
  !!! 寫在開頭的位置
  !!! 屬性名="value"
  a href="網址"
  img/video src="網址"

3. 內容: 顯示文字
  !!! 寫在開始和結束之間
```

網頁由三個東西構成(網頁前端):
1. 內容(main): HTML
2. 排版: CSS
3. 動態: JavsScript語法
  !!! 某個動作觸發了某些程式碼, 最後送出一個網址去得到更多東西

````python
# https://tabelog.com/tw/tokyo/rstLst/?SrtT=rt
# https://tabelog.com/tw/tokyo/rstLst/1/?SrtT=rt
import urllib.request as req
# beautifulsoup4
import bs4 as bs

url = "https://tabelog.com/tw/tokyo/rstLst/1/?SrtT=rt"
response = req.urlopen(url)
# content = response.read()
# json.loads做型態轉換
html = bs.BeautifulSoup(response)
# 區塊形態: 兩個專屬功能
# find 找第一個 find_all 找所有
# 參數1: html名字 參數2: 篩選條件
link = html.find("a", {"class":"list-rst__rst-name-target"})
print(link)
# 找到妳想要的區塊以後, 妳會想要萃取
# 1. 特殊屬性(字典): link["href"]
# 2. 內容(專屬值): link.text
# 程式語言基本概念: 一個形態可以擁有兩種東西
# a. 專屬功能: 形態.功能() -> 人.吃飯(1hour)
# b. 專屬值: 形態.值 -> 人.姓名
print(link["href"])
print(link.text)
````

````python
import urllib.request as req
import bs4 as bs

url = "https://tabelog.com/tw/tokyo/rstLst/1/?SrtT=rt"
response = req.urlopen(url)
html = bs.BeautifulSoup(response)

# find_all: 形態是個類似list的
r_list = html.find_all("div", {"class":"list-rst__body"})
for r in r_list:
    en_name = r.find("a", {"class":"list-rst__rst-name-target"})
    en_name_href = en_name["href"]
    en_name_text = en_name.text

    print(en_name_text)
    print(en_name_href)
    print("-" * 30)

# link_list = html.find_all("a", {"class":"list-rst__rst-name-target"})
# for link in link_list:
#     print(link["href"])
#     print(link.text)
#     print("-" * 30)
````
